```python
%matplotlib inline
from IPython.display import HTML, display
import time
import numpy as np
import pandas as pd
from product import Product
from product_transform import Transform
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = []
text_data = []
for product in Product().get_all():
    text = Transform(product).get('specification').refined().text().lower()
    data.append({'product_id': product['id'], 'text': text})

    text_data = pd.DataFrame(data)
```

```python
vec = TfidfVectorizer(analyzer='word', stop_words='english', min_df=2, ngram_range=(1,1))
tfidf_matrix = vec.fit_transform(text_data['text'])
```

```python
df = pd.DataFrame(tfidf_matrix.toarray(), index=text_data['text'], columns=vec.get_feature_names()).head(10)
display(df)
```

```python
vec.get_feature_names()
        
```

```python
count = 0
for results in tfidf_matrix.toarray():
    for value in results:
        if value <= 0:
            count += 1
print(count)
print(len(vec.get_feature_names()))
print(count / (len(vec.get_feature_names()) * 65))
```

```python
print(tfidf_matrix.toarray())
```

```python
cosine_similarities = cosine_similarity(tfidf_matrix, tfidf_matrix)
print(cosine_similarities)
```

```python
indices = np.argsort(vec.idf_)[::-1]
features = vec.get_feature_names()
top_n = 10
top_features = [features[i] for i in indices[:top_n]]

print(top_features)
```

```python
vec = TfidfVectorizer(analyzer='word', stop_words='english', min_df=2)
tfidf_matrix = vec.fit_transform(text_data['text'])

indices = np.argsort(vec.idf_)[::-1]
features = vec.get_feature_names()
top_n = 10
top_features = [features[i] for i in indices[:top_n]]

print(top_features)
```

```python
print('sparse matrix shape:', tfidf_matrix.shape)
print('nonzero count:', tfidf_matrix.nnz)
print(tfidf_matrix.shape[0] * tfidf_matrix.shape[1])
print('sparsity: %.2f%%' % (100.0 * tfidf_matrix.nnz / (tfidf_matrix.shape[0] * tfidf_matrix.shape[1])))

cvec = CountVectorizer(stop_words='english', min_df=2, ngram_range=(3,5))
cvec = cvec.fit_transform(text_data['text'])

print('sparse matrix shape:', cvec.shape)
print('nonzero count:', cvec.nnz)
print('sparsity: %.2f%%' % (100.0 * cvec.nnz / (cvec.shape[0] * cvec.shape[1])))
```

```python
weights = np.asarray(tfidf_matrix.sum(axis=0)).ravel().tolist()
weights_df = pd.DataFrame({'term': vec.get_feature_names(), 'weight': weights})
weights_df.sort_values(by='weight', ascending=False).head(20)
```

```python
vec = TfidfVectorizer(analyzer='word', stop_words='english', min_df=2, ngram_range=(1,3))
tfidf_matrix = vec.fit_transform(text_data['text'])

print('sparse matrix shape:', tfidf_matrix.shape)
print('nonzero count:', tfidf_matrix.nnz)
print('sparsity: %.2f%%' % (100.0 * tfidf_matrix.nnz / (tfidf_matrix.shape[0] * tfidf_matrix.shape[1])))

weights = np.asarray(tfidf_matrix.sum(axis=0)).ravel().tolist()
weights_df = pd.DataFrame({'term': vec.get_feature_names(), 'weight': weights})
weights_df.sort_values(by='weight', ascending=False).head(20)
```
